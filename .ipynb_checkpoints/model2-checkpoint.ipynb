```python
import numpy as np
import pandas as pd
import seaborn as sns
import time as t
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import LabelEncoder
from helper_functions import *
from sklearn.model_selection import train_test_split
from Pre_processing import *
#Read data from file #
data = pd.read_csv('player-value-prediction.csv')
#split column 'position'
data[['p1', 'p2', 'p3','p4']] = data['positions'].str.split(',', expand=True)
#Take columns which have null values less than 0.25
data = data.loc[:, data.isnull().sum() < 0.25*data.shape[0]]
#convert categories data to numbers

cols=['nationality','p1','preferred_foot','work_rate','body_type','club_team','club_position']
data=Feature_Encoder(data,cols)
#Replace null value with mean at wage and value columns
value_mean=data["value"].mean()
wage_mean = data['wage'].mean()
data['value'].fillna(value=value_mean, inplace=True)
data['wage'].fillna(value=wage_mean, inplace=True)
data.dropna(how='any',inplace=True)
#Date encodig
date_toInt(data,'contract_end_year')
date_toInt(data,'birth_date')
date_toInt(data,'club_join_date')

Columns=['LS','ST','RS','LW','LF','CF','RF','RW','LAM','CAM','RAM','LM','LCM','CM','RCM','RM','LWB','LDM','CDM','RDM','RWB','LB','LCB','CB','RCB','RB']
for i in Columns:
    position(i,data)
#Drop columns which do not affect on our model

data=data.drop(['id','name','full_name','birth_date','positions'], axis = 1)


#Split data
X=data.iloc[:,:-1]
Y=data["value"]

#Get the correlation between the features
data=data.iloc[:,:]
corr = data.corr()
#Top 50% Correlation training features with the Value
top_feature = corr.index[abs(corr['value'])>0.46]
#Correlation plot
plt.subplots(figsize=(12, 8))
top_corr = data[top_feature].corr()
sns.heatmap(top_corr, annot=True)
plt.show()
top_feature = top_feature.delete(-1)
X = X[top_feature] 
#Drop features that have high correlatin with other features 
X=X.drop(['reactions','composure','CM','RCM','ST','RS','CF','RF','CAM','RAM'], axis = 1)
#Feature Selection
X=featureScaling(X,1,100)

time1 = t.time()
#Split the data to training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.20,shuffle=True,random_state=10)
cls = linear_model.LinearRegression()
cls.fit(X_train,y_train)
prediction= cls.predict(X_test)
print('Co-efficient of linear regression',cls.coef_)
print('Intercept of linear regression model',cls.intercept_)
print('Mean Square Error', metrics.mean_squared_error(np.asarray(y_test), prediction))
time2 = t.time()
print('training time ',time2-time1)
M_L_R = linear_model.LinearRegression()
M_L_R.fit(X_train, y_train)
prediction = M_L_R.predict(X_test)
prediction = M_L_R.predict(X_test)
print('accuracy = ', M_L_R.score(X_test, y_test)*100)
```
